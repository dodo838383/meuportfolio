```python
#leitura dos dados
export_data = pd.read_csv('/content/EXP_1997_2020_20220808.csv',sep=';')
```

```python
export_data.head()
```

```python
uf_cod = {
"Acre": "AC",
"Alagoas": "AL",
"Amapá": "AP",
"Amazonas": "AM",
"Bahia": "BA",
"Ceará": "CE",
"Distrito Federal": "DF",
"Espírito Santo": "ES",
"Goiás": "GO",
"Maranhão": "MA",
"Mato Grosso": "MT",
"Mato Grosso do Sul": "MS",
"Minas Gerais": "MG",
"Paraná": "PR",
"Paraíba": "PB",
"Pará": "PA",
"Pernambuco": "PE",
"Piauí": "PI",
"Rio de Janeiro": "RJ",
"Rio Grande do Norte": "RN",
"Rio Grande do Sul": "RS",
"Rondônia": "RO",
"Roraima": "RR",
"Santa Catarina": "SC",
"Sergipe": "SE",
"São Paulo": "SP",
"Tocantins": "TO",
"Não Declarada": "NA"
}
```

####**Renomear colunas**

```python
cols_to_rename = {
    'Ano': 'ano',
    'UF do Produto': 'UF',
    'Codigo SH2': 'SH2',
    'Descrição SH2': 'SH2_detalhe',
    'Valor FOB (US$)': 'exportacao',
    'Quilograma Líquido': 'kilogramas'
}

export_data.rename(columns = cols_to_rename, inplace = True)
export_data.head()
```

####**Criando coluna com o código do município**

```python
export_data['UF_code'] = export_data['UF'].map(uf_cod)
```

*   **Filtro da base para o ano de 2020**
*   **Cálculo da soma das exportações por estado** 




```python
exportadores_2020 = export_data\
.query('ano == 2020')\
.groupby(by = 'UF_code')['exportacao'].sum()\
.sort_values()
```

#####**Junção das bases**

```python
data = pd.merge(
    estados,
    exportadores_2020,
    left_on = 'abbrev_state',
    right_on = 'UF_code',
    how = 'inner'
)
```

```python
fig, ax = plt.subplots(figsize=(12,12))
data.plot(
    column = 'exportacao',
    cmap = 'Blues',
    ax = ax,
    legend = True,
    edgecolor = 'BLACK',
    legend_kwds = {
        'label': 'Exportação',
        'orientation': 'horizontal'
    },
    linewidth = 1.0 
)
ax.set_title('PRINCIPAIS EXPORTADORES BR em 2020',
             fontweight = 'bold',
             fontsize = 12)
ax.axis('OFF');
```
